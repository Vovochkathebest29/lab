```python
# импортируем пакеты 
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import animation 
```

```python
# временный массив 
t = np.linspace(0,20,100)

# Позиционные массивы
x = np.sin(np.pi/5 * t)
y = np.sin(np.pi/3 * t)
z = np.linspace(0,20,100)

# Задаем набор данных для анимации 
dataSet = np.array([x,y,z])
# Комбинируем наши позиционные координаты 
numDataPoints = len(t)
```

```python
def animate_func(num):
  ax.clear() # Очищаем фигуру для обновления линии, точки,
  # Обновляем линии траектории (num+1 из-за индексации Python)
  ax.plot3D(dataSet[0, :num+1], dataSet[1, :num+1],
           dataSet[2, :num+1], c='blue')
  # Обновляем локации точки 
  ax.scatter(dataSet[0, num],dataSet[1, num],dataSet[2, num],
            c='blue', marker='o')
  # Добавляем постоянную начальную точку
  ax.plot3D(dataSet[0, :0], dataSet[1, :0],
           dataSet[2, :0], c='blue')
  # Задаем пределы для осей 
  ax.set_xlim3d([-1,1])
  ax.set_ylim3d([-1,1])
  ax.set_zlim3d([0,100])


  # Добавляем метки 
  ax.set_title('Trajectory \nTime = '+ str(np.round(t[num], decimals=2))+ 'sec')
  ax.set_xlabel('x')
  ax.set_ylabel('y')
  ax.set_zlabel('z')
 
```

```python
# Рисуем анимацию 
fig = plt.figure()
ax = plt.axes(projection='3d')
line_ani = animation.FuncAnimation(fig, animate_func, interval=100, frames=numDataPoints)
plt.show()
```

```python
# Сохраняем анимацию 
f = r'/content/animate_func.gif'
writegif = animation.PillowWriter(fps=numDataPoints/6)
line_ani.save(f, writer=writegif)

```

```python
import math
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# параметры задачи
z_speed = 5  # скорость зайца
d_speed = 5.35  # скорость собаки
radius = 5   # радиус круглой площадки

# задаем начальные координаты зайца и собаки
z_start = (0, radius)
d_start = (radius, 0)

# функция для вычисления дистанции между двумя точками
def distance(p1, p2):
    return math.sqrt((p1[0]-p2[0])**2 + (p1[1]-p2[1])**2)

# функция для анимации
def animate(i):
    # вычисляем текущие координаты зайца и собаки
    z_angle = (i * z_speed) % 360
    z_pos = (math.sin(math.radians(z_angle)) * radius, math.cos(math.radians(z_angle)) * radius)
    d_pos = d_start
    
    # если собака еще не догнала зайца
    if distance(z_pos, d_pos) > d_speed:
        # вычисляем угол между вектором движения собаки и вектором на зайца
        d_angle = math.degrees(math.atan2(z_pos[1]-d_pos[1], z_pos[0]-d_pos[0]))
        # вычисляем новые координаты собаки
        d_pos = (d_pos[0] + math.cos(math.radians(d_angle)) * d_speed,
                 d_pos[1] + math.sin(math.radians(d_angle)) * d_speed)

    # рисуем текущее положение зайца и собаки
    plt.cla()
    plt.gca().set_aspect('equal', adjustable='box')
    plt.xlim(-radius-1, radius+1)
    plt.ylim(-radius-1, radius+1)
    plt.plot([z_pos[0], d_pos[0]], [z_pos[1], d_pos[1]])
    plt.plot(z_pos[0], z_pos[1], 'o', color='gray')
    plt.plot(d_pos[0], d_pos[1], 'o', color='brown')

# создаем анимацию
fig = plt.figure()
anim = FuncAnimation(fig, animate, frames=360//z_speed*10, interval=50)
anim.save('погоня собаки за зайцем.mp4', fps=30, extra_args=['-vcodec', 'libx264'])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

R = 700
Vhare = 5
Vdog = 5.35
dt = 1
da = (Vhare*dt) / R

n = 800  # Количество кадров анимации

Xhare = np.zeros(n)
Yhare = np.zeros(n)
Xdog = np.zeros(n)
Ydog = np.zeros(n)
f = np.zeros(n)

Xdog[0] = -R
Ydog[0] = 0

for i in range(n):
    Xhare[i] = R * np.sin(da*i)
    Yhare[i] = R * np.cos(da*i)
    f[i] = np.arctan2((Yhare[i] - Ydog[i]), (Xhare[i] - Xdog[i]))

fig, ax = plt.subplots()
ax.set_xlim(-R, R)
ax.set_ylim(-R, R)

hare_line, = ax.plot([], [], 'b-', label='Hare')
dog_line, = ax.plot([], [], 'y-', label='Dog')
hare_point, = ax.plot([], [], 'b')
dog_point, = ax.plot([], [], 'y')

def init():
    hare_line.set_data([], [])
    dog_line.set_data([], [])
    hare_point.set_data([], [])
    dog_point.set_data([], [])
    return hare_line, dog_line, hare_point, dog_point

def animate(i):
    hare_line.set_data(Xhare[:i+1], Yhare[:i+1])
    dog_line.set_data(Xdog[:i+1], Ydog[:i+1])
    hare_point.set_data(Xhare[i], Yhare[i])
    dog_point.set_data(Xdog[i], Ydog[i])
    return hare_line, dog_line, hare_point, dog_point

def update(frame):
    global Xdog, Ydog, f
    if frame < n-1:
        Xdog[frame+1] = Xdog[frame] + Vdog * dt * np.cos(f[frame])
        Ydog[frame+1] = Ydog[frame] + Vdog * dt * np.sin(f[frame])
        f[frame+1] = np.arctan2((Yhare[frame] - Ydog[frame]), (Xhare[frame] - Xdog[frame]))

    return animate(frame)

ax.legend()

ani = animation.FuncAnimation(fig, update, frames=n, interval=100, blit=True, init_func=init)
ani.save('double.mp4',fps=30, extra_args=['-vcodec','libx264'])
plt.show()
```
